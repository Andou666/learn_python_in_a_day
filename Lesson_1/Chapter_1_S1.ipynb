## プログラムはコンピュータへの指令書
- 「**プログラム**」は機械に「こう動け」と命令する指令書
- PCのアプリだけでなく、ブラウザのサービス（検索エンジンやウェブメールなど）も提供会社側でプログラムが動いている
- プログラムを書くことを「**プログラミング**」という
- プログラムのことを「**コード**」、プログラミングのことを「**コーディング**」ともいう

## プログラミング言語で機械への命令が簡単に
- 機械は電気のON/OFF（01の2進数）で状態を管理している
- 最終的に01で命令を機会に伝える必要があるが、人間が01を扱うことは難しい
- プログラミング言語は「01しか使えない機械」と「自然で曖昧な言葉を使う人間」の間にある「厳格な文法の人工言語」
- 人間がプログラミング言語で命令を書き、機械がそれを01に変換して解釈実行する

## Pythonを選ぶ理由
- 「なにをやりたいか（目的）」に応じて使うプログラミング言語は変わってくる
- Pythonは「簡単である（すぐに覚えられる、同じプログラムを他の言語より素早くかける）」ことに加えて、「広範囲の開発作業に使える」「インタプリタ型」という特徴があるのでおすすめ

|*プログラミング言語*|簡単さ|速度|汎用性|実行方式|
|---|---|---|---|---|
|Python|◎|△|◎|インタプリタ（OSに依存しない）|
|Java|〇|〇|〇|コンパイル（OSに依存しない）|
|Python|△|◎|△|コンパイル（OSに依存する）|


## コンパイル型とインタプリタ型の言語の違い
- コンパイル型：プログラミング言語から機械語（01）に事前変換
- インタプリタ型：実行時にプログラミング言語から機械語に変換
- Pythonはインタプリタ型の言語
- インタプリタ型はコンパイルの手間（時間）がないので開発が楽

## Pythonが苦手なこと
- 超高速性が求められるシステムやデスクトップアプリケーション（一般ユーザ向けの再配布）、組込機器の開発が苦手。多くのスクリプト系言語も全く同じ領域が苦手
- 高速な処理（科学計算系はライブラリが高速なので問題なし）。一般的にインタプリタ型はコンパイラ型に比べると実行時にコンパイルするので遅い。速度は言語より設計やアルゴリズムに起因することが多いので、言語の速度はプログラミング上級者になってから気にすれば問題なし
- Pythonがインストールされている環境でPythonアプリを利用することは簡単だが、Pythonがインストールされていない環境（一般ユーザのPC）で動かすの不可能ではないものの手間がかかる
- 組込機器のような機会に近いレイアはPythonが使えないことが多い。基本は汎用OSの上でPythonを使う

## 汎用性の高さもPythonの魅力
- プログラミング言語には得意苦手がある
- 様々なことが得意な言語を学べば何にでもその言語を使える
- 利用目的が限られる言語は目的ごとに複数言語の習得が必要
- C言語：OSに近い領域や組込向けのプログラミングが得意
- Python
    - 使い捨ての短い簡単なスクリプトでも使える
    - 大規模なシステム開発でも使える
    - ウェブサービスの開発でも使える
    - 科学計算（機械学習/ビッグデータ処理など）でも使える

## プログラミング言語の分類とPython
- プログラミング言語は以下の三種類に分類できる
    - 手続き型言語（オブジェクト指向型言語のベース）
    - オブジェクト指向型言語（現在の主流。Pythonも）
    - 関数型言語
- ただしPythonは手続き型のような書き方もできるし、関数型のような書き方（中上級者向け）もできる
